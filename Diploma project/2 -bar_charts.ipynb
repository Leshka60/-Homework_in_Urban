**Столбчатая диаграмма** — это график в виде вертикальных или горизонтальных столбцов, высота или длина каждого из которых пропорциональна \
соответствующим им значениям.
С помощью столбчатых диаграмм удобно визуально оформлять категориальные данные — информацию, распределённую по дискретным группам (категориям). \
Например, возрастные группы людей, месяцы, размеры, виды животных и т. д..

1. Построение столбчатой диаграммы с использованием библиотеки **Matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO


# Создание DataFrame
df = pd.read_csv(StringIO(data))

# Преобразование столбца 'Date' в формат datetime
df['Date'] = pd.to_datetime(df['Date'])

# Установка индекса для DataFrame
df.set_index('Date', inplace=True)

# Построение столбчатой диаграммы
df.plot(kind='bar', figsize=(12, 6))

# Настройка графика
plt.title('Доля операционных систем с ноября 2023 по ноябрь 2024')
plt.xlabel('Дата')
plt.ylabel('Доля (%)')
plt.xticks(rotation=30)
plt.legend(title='Операционные системы')
plt.gca().legend(loc='upper left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()

# Отображение графика
plt.show()
```

2. Построение столбчатой диаграммы с использованием библиотеки **Seaborn**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from io import StringIO


# Создание DataFrame
df = pd.read_csv(StringIO(data))

# Преобразование столбца 'Date' в формат datetime
df['Date'] = pd.to_datetime(df['Date'])

# Преобразование DataFrame в длинный формат для Seaborn
df_melted = df.melt(id_vars='Date', var_name='OS', value_name='Share')

# Построение столбчатой диаграммы с использованием Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='Date', y='Share', hue='OS', errorbar=None)

# Настройка графика
plt.title('Доля операционных систем с ноября 2023 по ноябрь 2024')
plt.xlabel('Дата')
plt.ylabel('Доля (%)')
plt.xticks(rotation=45)
plt.legend(title='Операционные системы')
plt.gca().legend(loc='upper left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()

# Отображение графика
plt.show()
```
