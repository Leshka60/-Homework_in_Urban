**Круговая диаграмма** — это способ представления данных в виде круга, разделённого на сектора. Каждый сектор — это категория данных, \
которая составляет долю от общей суммы. Главная цель круговой диаграммы — показать, из чего состоит круг, то есть какие доли внутри \
единого целого занимают отдельные элементы. Поэтому сумма значений всех секторов должна составлять 100%.

1. Построение круговой диаграммы с использованием библиотеки **Matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO


# Создание DataFrame
df = pd.read_csv(StringIO(data))

# Выбор данных для круговой диаграммы (например, для последнего месяца)
latest_data = df.iloc[-1].drop('Date')

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(latest_data, labels=latest_data.index, autopct='%1.1f%%', startangle=50)

# Настройка графика
plt.title('Доля операционных систем в ноябре 2024 года')
plt.axis('equal')  # Для равного соотношения осей
plt.tight_layout()

# Отображение графика
plt.show()
```

2. Построение круговой диаграммы с использованием библиотеки **Seaborn**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from io import StringIO


# Создание DataFrame
df = pd.read_csv(StringIO(data))

# Выбор данных для круговой диаграммы (например, для последнего месяца)
latest_data = df.iloc[-1].drop('Date')

# Установка стиля Seaborn
sns.set(style="whitegrid")

colors = sns.color_palette('bright')

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(latest_data, labels=latest_data.index, autopct='%1.1f%%', startangle=50, colors=colors)

# Настройка графика
plt.title('Доля операционных систем в ноябре 2024 года')
plt.axis('equal')  # Для равного соотношения осей
plt.tight_layout()

# Отображение графика
plt.show()
```
